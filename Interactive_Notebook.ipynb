#REPETIÇÃO

  Neste capitulo vamos comentar do for e do while, que podem ser consideradas

**FOR**

Para começarmos e darmos um exemplo vamos gerar uma lista com 3 valores e atraves do "For" e percorre-la mostrando na tela os valores.

Sua sintaxe básica é: for **<nome variável>** in **<iterável>**. Vamos entender:

````python
#Lista com tres valores.
lista=[1,2,3]
# Exemplo que mostra na tela os itens da lista.
for i in lista:
  print(lista[i])
````

Esse exemplo de Codigo indicou para o "for" que ele deve percorrer a lista e seu contador vai ser a variavel "i". Dentro do for há um "print(i)", que irá mostrar na tela o item da lista referente a posição "i" cada ver que o "for" rodar.

Vamos comentar agora de uma estrutura que é extremamente utilizada juntamente com o "for" O "Range" que nos permite indicar quantas vezes desejamos que o "for" venha a se repetir.
Para começarmos vamos gerar uma lista com 5 valores e atraves do "For" vamos percorrer apenas 3 dos valores da lista e mostra-los na tela utilizando a mesma estrategia do codigo anterior, utilizando o "print(i)".

````python
#Lista com cinco valores.
lista=[1,2,3,4,5]
# Exemplo que mostra na tela tres dos cinco itens da lista.
for i in range(3):
  print(lista[i])
````

Vale a pena comentar que existem auxiliadores que nos permite modificar a sequencia de um laço de repetição, podem ser utilizados tanto no "for" quanto no "while", são eles break, continue e pass.

O **break** é usado para finalizar um loop

O **pass** nada mais é que uma forma de fazer um código que não realiza operação nenhuma. É utilizado principalmente para darmos continuidade em um programa que seria encerrado por um codigo de erro.

O **Continue** Serve para indicar que apos uma verificação positiva o codigo deve ser continuado saindo do laço de repetição.


**WHILE**

Para começarmos e darmos um exemplo vamos gerar uma lista com 3 valores e atraves do "while" percorre-la mostrando na tela os valores.

É importante dizer que difente do "for", temos que realizar um trabalho mais minucioso em relação ao contador, devemos iniciar um variavel que iremos utilizar como o contador do "while". Dentro do mesmo laço devemos realizar uma soma a variavel contador, para que o laço tenha uma continuação até que o teste para entrar no while seja falso e o programa de continuidade.

Sua sintaxe é:

```
while "teste":
   codigo
"contador"+1
```


````python
#Lista com cinco valores.
lista=[1,2,3,4,5]
#Variavel que servirá como contador
i = 0
#Iniciamos o while com seu teste, testamos o contador, assim que o teste for falso o programa vai dar continuidade saindo do laço
while i < 5:
  print(lista[i])
  #Somando um valor ao contador
  i += 1
````

**Vetor**

Vetores são estruturas e facilitam a armazenagem e manipulação de um conjunto de dados. Em python os vetores utilizam a estrutura de 'listas'('list').

Uma lista em
Python é definida da seguinte forma:

```
Nome_Lista = [ valor1, valor2, ..., valorN]
```

Por	exemplo,	para	alocar	um	vetor	usamos	o	comando:

```
vetor = [5];
```
Que	irá	alocar	o	espaço	de	5	valores	inteiros	na	memória:

vetor.PNG

Perceba	que	o	vetor	foi	criado	sem	nenhum	valor.	Na	prática,	será	criado	esse	vetor
com	valores	aleatórios	que	estão	armazendo	na	posição	da	memória.

No exemplo vamos criar um vetor chamado dias e  atribuímos os dias da semana como conteúdo do vetor.

```
dias = ["domingo", "segunda", "terça", "quarta", "quinta", "sexta", "sabado"]
```

Para acessar cada dia da semana, usamos a representação dias[i] onde i representa o índice do elemento do vetor. No caso da linguagem Python o primeiro elemento de um vetor tem i=0, logo dias[0] representa "domingo" e dias[1] "segunda" e assim sucessivamente. Por exemplo para que o programa imprima o valor "segunda".

```
dias = ["domingo", "segunda", "terça", "quarta", "quinta", "sexta", "sabado"]
print(dias[1])
```

* Podemos adicionar também variáveis em cada índice de um vetor.

```

```

**Matriz**

Em Python, uma matriz pode ser representada como uma lista de listas, onde um elemento da lista contém uma linha da matriz, que por sua vez corresponde a uma lista com os elementos da coluna da matriz.

matriz.PNG

Para entender melhor, vamos fazer 3 vetores, e inserir cada vetor em um índice de um novo vetor:

```
vetor1 = ["Bom Dia", "Boa Tarde", "Boa Noite"]
vetor2 = ["Bruno", "Marcelo", "Roberto"]
vetor3 = ["Como Vai?", "Que horas são?","Vamos programar?"]
matriz = [vetor1, vetor2, vetor3]
```

Cada vetor se comportará como uma coluna, e cada índice da matriz representará uma coluna. Ou seja: A primeira coluna da matriz (índice 0) será o 'vetor1', a segunda (índice 1) o 'vetor2' e terceira (índice 2) 'vetor3'.

* Podemos também usar o método *append* para inserir cada vetor:
```

matriz = []

matriz.append(vetor2)
matriz.append(vetor3)
print(matriz[1])
  ['Bruno', 'Marcelo', 'Roberto']
```
